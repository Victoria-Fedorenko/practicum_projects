## Сегментация

Изначально попробуем решить задачу по сегментации клиентов с использованием машинного обучения посредством кластеризации. 

```python
# преобразуем категориальные переменные в числовые
clients_ml = pd.get_dummies(clients)
clients_ml.head()
```

```python
scaler = StandardScaler()
x_sc = scaler.fit_transform(clients_ml)
```

Чтобы определить, какое количество кластеров будет подходящим, построим дендрограмму.

```python
linked = linkage(x_sc, method = 'ward') 

plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Кластеризация клиентов банка')
plt.show() 
```

На дендрограмме видим **5 кластеров**. 

```python
km = KMeans(n_clusters=5, random_state=0) 
labels = km.fit_predict(x_sc)
clients_ml['clusters'] = labels
clients_ml.groupby(['clusters']).mean()
```

Кластеры, которые сформировалиьсь в результате машинного обучения, не удовлетовряют запрос нашего заказчика. Так, количество потребляемых продуктов одинакво в каждом из кластеров. Видим при этом, что выделенные кластеры обращают наше внимание на другие аспекты выборки. Кластеры отличаются по городам, уровню дохода и среднему оттоку. В исследовательском анализе данных мы уже обратили внимание на эти особенности. 

```python
print(len(clients.query('products > 2 and churn == 1')) / len(clients.query('products > 2')))
```

По информации от заказчика, в сегментации ключевым признаком клиента для нас является количество потребляемых им продуктов. Вторым по важности является отток. Проведем сегментацию самостоятельно, опираясь на эти признаки. 

Предлагаем следующие сегменты для анализа:

**segment_1**: 1 используемый продукт, клиент ушел.

**segment_2**: 1 используемый продукт, клиент не ушел.

**segment_3**: 2 и более используемых продуктов, клиент не ушел. 

**segment_4**: 2 используемых продукта, клиент ушел. 

**segment_5**: Более 2 используемых продуктов, клиент ушел. 

Среди клиентов, которые используют более двух продуктов, очень многие перестают быть клиентами банка, поэтому кажется важным рассмотреть этот сегмент отдельно. 

```python
def client_type(products, churn):
    try:
        if products == 1:
            if churn == 1:
                return 'segment_1'
            else:
                return 'segment_2'
        if products >= 2:
            if products == 2:
                if churn == 1:
                    return 'segment_4'
                else:
                    return 'segment_3'
            if products > 2:
                if churn == 1:
                    return 'segment_5'
                else:
                    return 'segment_3'
    except:
        return 'ошибка'
    

    
clients['client_type'] = clients.apply(lambda x: client_type(x['products'], x['churn']), axis = 1)
clients.head()
```

```python
# проверяем, возникли ли ошибки в сегментации
print(len(clients.query('client_type == "ошибка"')))
```

```python
clients.groupby('client_type')['user_id'].count().sort_values(ascending=False)
```

Больше всего пользователей относятся к третьему сегменту, и чуть меньше - ко второму. Оба эти сегмента - группы пользователей, которые остаются клиентами нашего банка, и отличаются они количеством используемых продуктов. В третьем сегменте речь идет о двух и более продуктах, а во втором сегменте - об одном продукте. В то же время среди клиентов, которые уходят, больше всего тех, кто использует только один продукт (первый сегмент).

```python
clients.groupby('client_type').agg(
    {'score':'mean', 'age':['mean', 'median'], 'objects_in_property':'mean', 
     'account_balance':['mean', 'median'], 'credit_card':'mean', 'salary':['mean', 'median'], 'loyalty':'mean'})
```

```python
clients.groupby('client_type')[['city']].agg(pd.Series.mode)
```

Таблицы показывают, что средний возраст клиентов в сегментах, которым свойственен отток, примерно на 8 лет больше, чем в сегментах, которые остаются клиентами банка. Кажется, что нет значимого различия в зарплатах, доле пользователей кредитных карт,  среднем скоринговом балле. Примечательно, что в третьем сегменте - сегменте пользователей, которые используют более 2 продуктов банка и остаются его пользователями - самое маленькое среднее и медианное значение баланса на счете (для половины пользователей у счета нулевой баланс) при сопоставимым с другими сегментами уровнем зарплаты. Предполагаем, что это может быть связано с более активным использованием кредитных карт или других кредитных продуктов.

В первом сегменте, самом большом сегменте среди "уходящих", самым частым значением города является Ростов Великий. Ранее мы замечали, что доля оттоков в этом городе больше, чем в двух других. Во всех остальных сегментах преобладают клиенты из Ярославля - самого крупного города в выборке.

Что касается пятого сегмента, клиенты в котором уходят и при этом используют более двух продуктов, но его показатели в целом соотносятся с показателями остальных сегментов. ## Статистическая проверка гипотез

Проверим следующие утверждения:

1. Доход клиентов, которые пользуются двумя продуктами банка, отличается от дохода клиентов, которые пользуются одним продуктом банка. 

2. Возраст клиентов, которые перестают пользоваться услугами банка, отличается от возраста клиентов, которые остаются клиентами банка.

### Проверка гипотезы о доходах

Для начала проверим, является ли распределение нормальным. 

Нулевая гипотеза: распределение является нормальным.

Альтернативная гипотеза: распределение не является нормальным.

```python
salary = clients['salary']

alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(salary)
p_value = results[
    1
]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)


if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу: распределение является нормальным') 
```

Теперь сформулируем гипотезы о доходах.

Нулевая гипотеза: Доход клиентов, которые пользуются двумя продуктами банка, не отличается от дохода клиентов, которые пользуются одним продуктом банка. 

Альтернативная гипотеза: Доход клиентов, которые пользуются двумя продуктами банка, отличается от дохода клиентов, которые пользуются одним продуктом банка. 

Поскольку распределение не является нормальным, будем проверять гипотезу о доходах через U-тест. 

```python
salary_1_product = clients.query('products == 1')['salary']
salary_2_product = clients.query('products == 2')['salary']

alpha = 0.05

results = st.mannwhitneyu(salary_1_product, salary_2_product)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Таким образом, можно сделать вывод, что нет статистически значимого различия в доходах клиентов, использующих один продукт, и в доходах клиентов, использующих два продукта.

### Проверка гипотезы о возрасте

Для начала проверим, является ли распределение нормальным. 

Нулевая гипотеза: распределение является нормальным.

Альтернативная гипотеза: распределение не является нормальным.

```python
age = clients['age']

alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(age)
p_value = results[
    1
]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу: распределение является нормальным') 
```

Теперь сформулируем статистическую гипотезу о возрасте.

Нулевая гипотеза: Средний возраст клиентов, которые перестают пользоваться услугами банка, не отличается от среднего возраста клиентов, которые остаются клиентами банка.

Альтернативная гипотеза: Возраст клиентов, которые перестают пользоваться услугами банка, отличается от возраста клиентов, которые остаются клиентами банка.

Поскольку распределение не является нормальным, будем проверять гипотезу о доходах через U-тест. 

```python
churn_0_age = clients.query('churn == 0')['age']
churn_1_age = clients.query('churn == 1')['age']

alpha = 0.05

results = st.mannwhitneyu(churn_0_age, churn_1_age)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Делаем вывод о том, что есть статистическая значимость в разнице возрастов. Исследовательский анализ данных также свидетельствовал о том, что отток связан с возрастом.
